```python
import requests
import pandas as pd

url = "https://bank.gov.ua/NBU_Exchange/exchange_site?start=20220115&end=20220131&valcode=usd&sort=exchangedate&order=desc&json"
response = requests.get(url)
data = response.json()
dollars = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt

plt.plot(list(dollars['calcdate']), list(dollars['rate_per_unit']))
plt.xticks(rotation=90)
plt.xlabel('data')
plt.ylabel('rate per unit')
plt.title('rate of dollar')
plt.savefig('dolars.png')
```
